```python
import pandas as pd
import numpy as np
```

```python
heights = pd.read_csv('height.csv')
```

```python
#method chaining

def tweak_heights(df):
    return(heights
   .set_index('order')
   .assign(name = heights
           .name
           .astype('string')
           .str
           .lower()
          )
    .sort_values(
           by=["height(cm)"],
           ignore_index= True
         )
            )
```

```python
tweak_heights(heights)
```

```python
heights.describe()
```

```python
height = np.array(heights["height(cm)"])
print('mean height=',height.mean())
print('median height=',np.median(height))
print('minimum height=',height.min())
print('maximum height=',height.max())
print('standard deviation=',height.std())
print('25th percentile=',np.percentile(height,25))
print('75th percentile=',np.percentile(height,75))
```

```python
heights["height(cm)"].plot(kind='box')
```
